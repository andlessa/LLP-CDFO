```python
import pandas as pd
```

```python
recast_2018_0j = pd.read_pickle('./pp2BB0j_2018_13.pcl')
recast_2018_0j['Mode'] = '0 jet'
recast_2018_1j = pd.read_pickle('./pp2BB1j_2018_13_small_qcut.pcl')
recast_2018_1j['Mode'] = '1 jet (small qcut)'
recast_2018_1jb = pd.read_pickle('./pp2BB1j_2018_13.pcl')
recast_2018_1jb['Mode'] = '1 jet'
recast_2018 = pd.concat([recast_2018_0j,recast_2018_1j,recast_2018_1jb])
recast_2018.sort_values(['mLLP','mLSP','tau_ns','SR', 'Mode'],inplace=True,ignore_index=True)        

recast_2018 = recast_2018[recast_2018['SR'] == 'Trackless']
```

```python
recast_2018[['Mode','SR','$N_s$','AccEff','mLLP','mLSP','tau_ns','robs','rexp','Total xsec (pb)']].T
```
