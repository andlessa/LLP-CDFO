```python
import numpy as np
from matplotlib import pyplot as plt
from itertools import product
```

```python
mN1 = np.arange(100.,1500.,100.)
deltaM = np.arange(10.,35.,5.)
tau_cm = np.arange(1.,25.,5.)
tau = tau_cm/30.
pts = np.array(list(product(mN1,deltaM,tau)))
plt.scatter(pts[:,0],pts[:,1])
# plt.xscale('log')
# plt.xlim(1e-3,1e2)
# plt.ylim(25.,2000.)
plt.ylabel(r'$\Delta m$ (GeV)')
plt.xlabel(r'$m_{\tilde \chi_1^0}$ (GeV)')
plt.title(r'$N = %i$' %len(pts))
plt.show()
```

```python
ptsIni = [("%1.0f" %pt[0], "%1.1f" %pt[1], "%1.1e" %pt[2]) for pt in pts]
print(str(ptsIni).replace("'", ""))
```
